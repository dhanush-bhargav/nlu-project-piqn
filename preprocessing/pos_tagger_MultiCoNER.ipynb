```python
from flair.models import SequenceTagger
from flair.data import Sentence
from datasets import load_dataset
```

```python
pos_tagger = SequenceTagger.load("flair/upos-multi")
```

```python
dataset = load_dataset("MultiCoNER/multiconer_v2","German (DE)")
```

```python
print(dataset['train'][0])
```

```python
train_tokens = [example['tokens'] for example in dataset['train']]
validation_tokens = [example['tokens'] for example in dataset['validation']]
test_tokens = [example['tokens'] for example in dataset['test']]

print("Sample train tokens:", train_tokens[0])
```

```python
# POS Tagger
def pos_tag_tokens(token_lists):
    all_pos_tags = []

    for tokens in token_lists:
        sentence = Sentence(" ".join(tokens))
        pos_tagger.predict(sentence)
        pos_tags = [token.get_label("upos").value for token in sentence]
        all_pos_tags.append(pos_tags)

    return all_pos_tags
```

```python
train_pos_tags = pos_tag_tokens(train_tokens)
# validation_pos_tags = pos_tag_tokens(validation_tokens)
# test_pos_tags = pos_tag_tokens(test_tokens)
```

```python
for tokens, pos_tags in zip(train_tokens, train_pos_tags):
    print(f"Tokens: {tokens}")
    print(f"POS Tags: {pos_tags}")
```

```python
def create_token_pos_dict(tokens_list, pos_tags_list):
    """
    Creates a dictionary that maps tokens to their corresponding POS tags.

    Args:
        tokens_list (list of list of str): A list where each element is a list of tokens (a sentence).
        pos_tags_list (list of list of str): A list where each element is a list of POS tags (for a sentence).

    Returns:
        list of dict: A list where each element is a dictionary with 'tokens' and 'pos_tags'.
    """
    token_pos_dicts = []

    for tokens, pos_tags in zip(tokens_list, pos_tags_list):
        token_pos_dict = {
            "tokens": tokens,
            "pos_tags": pos_tags
        }
        token_pos_dicts.append(token_pos_dict)

    return token_pos_dicts
```

```python
train_dict = create_token_pos_dict(train_tokens, train_pos_tags)
```
